```python
def DataWafer350():
    _data = {}
    refl = []
    wl = []
    localpath = os.path.dirname(abspath(getsourcefile(lambda:0)))
    files = ['refl/25/Ag_film_refl_lamp_50um_350_650nm_0.dat', 'refl/26/Ag_refl_lamp_50um_350_650nm_0.dat', 'refl/25/SiSiO2_refl_lamp_50um_350_650nm_0.dat', 'refl/26/SiSiO2_refl_lamp_50um_350_650nm_0.dat', 'refl/26/Background_refl_lamp_50um_350_650nm_0.dat']
    for i, file in enumerate(files):
        file1 = os.path.join(localpath, file)
        data = np.genfromtxt(file1, skip_header=1, unpack=True)
        refl.append(data[1, :]) 
        wl = data[0, :]
    return wl, refl[0], refl[1], refl[2], refl[3], refl[4]
```

```python
def Data26():
    wl_1, Ag25_1, Ag26_1, S25_1, S26_1, Bg_1 = DataWafer350()
    wl_2, Ag25_2, Ag26_2, S25_2, S26_2, Bg_2 = DataWafer600()
    wl = np.hstack([wl_1, wl_2])
    rr = (np.hstack([S26_1, S26_2])/2 - np.hstack([Bg_1, Bg_2]))/(np.hstack([Ag26_1, Ag26_2])- np.hstack([Bg_1, Bg_2]))
    return wl, rr
```
